```python
from transformers import AutoTokenizer
from pyvi import ViTokenizer
from underthesea import word_tokenize
from vncorenlp import VnCoreNLP
```

```python
vncorenlp_init = VnCoreNLP(jar_path, annotators="wseg", max_heap_size='-Xmx2g')
```

```python
sample_text = "Học sinh học sinh học tại Thành phố Hồ Chí Minh."
```

```python
sample_text_pyvi = ViTokenizer.tokenize(sample_text)
print(sample_text_pyvi)
```

```python
sample_text_underthesea = word_tokenize(sample_text)
print(sample_text_underthesea)
```

```python
sample_text_vncorenlp = vncorenlp_init.tokenize(sample_text)
print(sample_text_vncorenlp)
```

```python
from IPython.display import clear_output

models_multilingual = [
    "Qwen/Qwen3-Embedding-0.6B",
    "microsoft/harrier-oss-v1-0.6b",
    "microsoft/harrier-oss-v1-270m",
    "Qwen/Qwen3-Embedding-0.6B",
    "google/embeddinggemma-300m",
    "Alibaba-NLP/gte-multilingual-base",
    "intfloat/multilingual-e5-large-instruct",
    "intfloat/multilingual-e5-large",
    "BAAI/bge-m3",

]

models_vietnamese = [
    "bkai-foundation-models/vietnamese-bi-encoder",
    "vinai/phobert-base",
    "vinai/phobert-large",
    "vinai/phobert-base-v2",
]

models = models_multilingual + models_vietnamese

# load models
for model_name in models:
    model = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    print(f"Loaded model: {model_name}")
    clear_output()
    
```

```python
# raw text
print(sample_text)
for model_name in models:
    model = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    tokens = model.tokenize(sample_text)
    print(f"{model_name}: {tokens}")
```

```python
# pre-tokenized text (using PyVi)
print(sample_text_pyvi)
for model_name in models:
    tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    pretokenized_text = sample_text_pyvi
    tokens = tokenizer.tokenize(pretokenized_text)
    print(f"Tokens for {model_name}: {tokens}")
```

```python
# pre-tokenized text (using underthesea)
pretokenized_text = " ".join(["_".join(word.split()) for word in sample_text_underthesea])
print(pretokenized_text)
for model_name in models:
    tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    tokens = tokenizer.tokenize(pretokenized_text)
    print(f"Tokens for {model_name}: {tokens}")
```

```python
# pre-tokenized text (using vncorenlp)
pretokenized_text = " ".join(sample_text_vncorenlp[0])
print(pretokenized_text)
for model_name in models:
    tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
    tokens = tokenizer.tokenize(pretokenized_text)
    print(f"Tokens for {model_name}: {tokens}")
```
